```python
import pandas as pd

summary_df = pd.read_csv("data/fatigue_summary.csv")
rotation_df = pd.read_csv("data/rotation_plan.csv")
```

```python
from IPython.display import display, Markdown

# Add explanation
display(Markdown(f"""
### 🧠 Fatigue Summary (Past Month)

Fatigue score = `total_alerts + 2 × night_alerts`  
"""))

# Style the DataFrame
styled_df = (
    summary_df.style
    .format({
        "avg_fatigue_score": "{:.1f}",
        "max_fatigue_score": "{:.0f}",
        "days_above_threshold": "{:.0f}"
    })
    .set_properties(**{"text-align": "center"})
)

display(styled_df)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))
summary_df.plot(
    x="engineer", 
    y=["avg_score", "max_score", "days_above_threshold"], 
    kind="bar", 
    ax=ax
)
plt.title("📊 Fatigue Metrics by Engineer")
plt.ylabel("Score / Days")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
